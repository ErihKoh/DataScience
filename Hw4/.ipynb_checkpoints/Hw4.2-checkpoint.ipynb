```python
from scipy.integrate import solve_ivp, odeint
import matplotlib.pyplot as plt
import numpy as np

α = 0.5
β = 0.3 
N = 1000000 
S = 990000 
I = 7000
R = 3000 
t0 = 0
tf = 25

def dSdt(t, y):
    return -α * y

result = solve_ivp(dSdt, [t0, tf], [S], dense_output=True)

t, y = result.t, result.y[0]

plt.plot(result.t, result.y[0])
z, = result.sol(t)
plt.plot(t, y, 'o', color='k', label='solve_ivp')

plt.plot(t, z, color='blue', label='Interpolation')
# plt.yticks(np.arange(0, N, step=100000))
plt.legend()
plt.show()
```

```python
def dRdt(t, y):
    return β * y

result2 = solve_ivp(dRdt, [t0, tf], [R], dense_output=True)

t2, y2 = result2.t, result2.y[0]
z2, = result2.sol(t2)
plt.plot(t2, y2, 'o', color='k', label='solve_ivp')
plt.plot(t2, z2, color='blue', label='Interpolation')
plt.legend()
plt.show()
```
