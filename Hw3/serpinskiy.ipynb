```python
import matplotlib.pyplot as plt
import numpy as np
import random

h = 1000
w = 1000
qnt = 15000

A = [0, 0]
B = [w/2, h]
C = [w, 0]

x = [A[0], B[0], C[0]]
y = [A[1], B[1], C[1]]


def random_dot(x, y):
    flag = True
    y0 = random.uniform(y[0], y[1])
    while flag:
        x0 = random.uniform(x[1], x[2])
        i = (x[0] - x0) * (y[1] - y[0]) - (x[1] - x[0]) * (y[0] - y0)
        j = (x[1] - x0) * (y[2] - y[1]) - (x[2] - x[1]) * (y[1] - y0)
        k = (x[2] - x0) * (y[0] - y[2]) - (x[0] - x[2]) * (y[2] - y0)
        
        if (i > 0 and j > 0 and k > 0) or (i < 0 and j < 0 and k < 0):
            
            flag = False
            return x0, y0
        


def dots():
    x0, y0 = random_dot(x, y) 
    next_point = [None, None]
    for i in range(15000):
        corner = random.choice([A, B, C])
        next_point[0] = (corner[0] + x0) / 2
        next_point[1] = (corner[1] + y0) / 2
        x.append(next_point[0])
        y.append(next_point[1])
        x0, y0 = next_point[0], next_point[1]
        
dots()     
        


plt.figure(figsize=(10, 10))

plt.scatter(x, y)
plt.show()
```
