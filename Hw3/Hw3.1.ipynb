```python
import pandas as pd
import requests
%matplotlib inline


response = requests.get(
    "https://uk.wikipedia.org/wiki/Населення_України#Народжуваність"
)

tmp = pd.read_html(response.text, thousands='.', decimal=',')
my_table = tmp[10]
```

```python
import matplotlib.pyplot as plt

years = my_table.columns[1::].astype(int)
ua = my_table.iloc[27][1::].astype(float)

plt.title('Коефіцієнт народжуваності в України (1950—2019)', fontsize=12)

plt.xlabel('Рік', fontsize='large', fontstyle='italic', color='midnightblue')
plt.ylabel('Коефіцієнт', fontsize='large', fontstyle='italic', color='midnightblue')
plt.xticks(years, rotation=90)


plt.plot(years, ua)
plt.grid(axis='y')

plt.show()

```

```python
fig, axs = plt.subplots(figsize=(15,5))
axs.bar(my_table['Регіон'], my_table['2019'], color=["b", "r", "y", "pink", "g", "m"])
plt.ylabel('Народжуваність на рік')
plt.title('Народжуваність в Україні за регіонами')
for tick in axs.get_xticklabels():
            tick.set_rotation(55)
plt.grid(axis='y')
plt.show()
```

```python
vinnica = my_table.iloc[1][1::].astype(float)
mykolaiv = my_table.iloc[13][1::].astype(float)
chernigiv = my_table.iloc[24][1::].astype(float)

plt.plot(years, vinnica)
plt.plot(years, mykolaiv)
plt.plot(years, chernigiv)

plt.title('Коефіцієнт народжуваності (1950—2019)', fontsize=12)

plt.xlabel('Рік', fontsize='large', fontstyle='italic', color='midnightblue')
plt.ylabel('Коефіцієнт', fontsize='large', fontstyle='italic', color='midnightblue')
plt.xticks(years, rotation=90)
plt.legend(['vinnica', 'mykolaiv', 'chernigiv'])
plt.grid(axis='y')

plt.show()
```
