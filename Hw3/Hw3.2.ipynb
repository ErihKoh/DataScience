```python
import matplotlib.pyplot as plt
import numpy as np
import random

x1 = 15
y1 = 25
h = 15

x = []
y = []

def dots_treangle(x, y, h):
    y2 = y3 = y - h * 0.866;
    x2 = x + h / 2
    x3 = x - h / 2 
    return x2, y2, x3, y3

def dot(y1, y2, x2, x3):
    flag = True
    y0 = random.uniform(y1, y2)
    while flag:
        x0 = random.uniform(x2, x3)
        i = (x1 - x0) * (y2 - y1) - (x2 - x1) * (y1 - y0)
        j = (x2 - x0) * (y3 - y2) - (x3 - x2) * (y2 - y0)
        k = (x3 - x0) * (y1 - y3) - (x1 - x3) * (y3 - y0)
        
        if (i > 0 and j > 0 and k > 0) or (i < 0 and j < 0 and k < 0):
            
            flag = False
            return x0, y0
        
def draw_treangle(x1, y1, h):
    
    x2, y2, x3, y3 = dots_treangle(x1, y1, h)   
    x0, y0 = dot(y1, y2, x2, x3)
    list_x = [x0, x1, x2, x3]
    list_y = [y0, y1, y2, y3]
    return list_x, list_y



x, y = draw_treangle(x1, y1, h)
        


plt.figure(figsize=(10, 10))

plt.scatter(x, y)
plt.xticks(range(1, 30, 1))
plt.yticks(range(1, 30, 1))
plt.show()
```
