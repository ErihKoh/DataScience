```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame(data=({'moments': [1, 2, 3, 4, 5, 6],'A': [25, -10, 10, 5, 35, 13], 'B': [0, 15, -5, 5, 20, 25], 'C': [10, 25, -15, -5, -5, 15]}))
```

```python
fig, ax = plt.subplots()
ax = sns.regplot(x = 'A', y = 'B', data=data)
```

```python
ax = sns.regplot(x = 'A', y = 'C', data=data)
```

```python
ax = sns.regplot(x = 'C', y = 'B', data=data)
```

```python
data[['A', 'B', 'C']].corr()
```
